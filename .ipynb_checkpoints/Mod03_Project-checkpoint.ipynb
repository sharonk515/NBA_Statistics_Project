# Player Data

## Creating function to get player data

```python
def get_player_data(per_page=100, total_pages=None):
    
    # Get the last page from the meta data
    url = 'https://www.balldontlie.io/api/v1/players'
    parameters = {'per_page': per_page, 'page': 1}
    resp = requests.get(url, params = parameters).json()
    total_pages = resp['meta']['total_pages'] if total_pages is None else total_pages
    
    # Initialize data frame 
    player_data = pd.DataFrame()
    
    start_time = time.time()
    
    # Loop to iterate through all pages
    for page in range(1, total_pages+1):
        
        # Making the request
        url = 'https://www.balldontlie.io/api/v1/players'
        parameters = {'per_page': per_page, 'page': page}
        resp = requests.get(url, params = parameters).json()

        # Creating dataframe from the data key
        data = pd.DataFrame.from_records(resp['data'])
        
        # Combine name
        data['full_name'] = data['first_name'] + " " + data['last_name']
        
        # Create total height column
        data['height_inches_total'] = data['height_feet']*12 + data['height_inches']

        # Creating columns out of the 'home_team' and 'visitor_team' column (these columns had dictionary values)
        for k in data['team'][0].keys():
            data['team_'+k] = data['team'].apply(lambda x: x[k])
        
        # Column to display which page results came from
        data['Page'] = page
        
        # Concatenate our initial datafram with dataframe from each loop
        player_data = pd.concat([player_data, data])
        
        # Delay between requests
        time.sleep(0.5)
    
    end_time = time.time()
    print(end_time - start_time)
    
    return player_data
```

# Games Data

## Creating function to request data 

```python
def get_games_data(per_page=None, total_pages=None):

    # Get the last page from the meta data
    url = 'https://www.balldontlie.io/api/v1/games'
    parameters = {'per_page': per_page, 'page': 1}
    resp = requests.get(url, params = parameters).json()
    total_pages = min(x for x in [resp['meta']['total_pages'], total_pages] if x is not None)
    
    # Initialize data frame 
    games_data = pd.DataFrame()
    
    
    # Loop to iterate through all pages
    for page in range(1, total_pages+1):
        
        # Specifyig parameters
        per_page = min(x for x in [per_page, 100] if x is not None)
        parameters = {'per_page': per_page, 'page': page}
        
        # Making the request
        resp = requests.get(url, params = parameters).json()

        # Creating dataframe from the data key
        data = pd.DataFrame.from_records(resp['data'])
        
        # Filter data to only show 'Final' (finished) games
        data = data.loc[data['status']=="Final"]
       
        # Creating columns out of the dictionary-type columns
        for col in data.columns:
            if type(data[col][0]) == dict:
                for k in data[col][0].keys():
                    data[col+"_"+k] = data[col].apply(lambda x: x[k])
        
        
        # Filter data to only show East vs West games
        data = data.loc[((data['home_conference']=="East") & (data['visitor_conference']=="West")) 
                        | ((data['home_conference']=="West") & (data['visitor_conference']=="East"))]
        
        # Column to display which page results came from
        data['page'] = page
        
        # Concatenate our initial datafram with dataframe from each loop
        games_data = pd.concat([games_data, data])
        
        # Delay between requests
        time.sleep(0.8)
    
    # Convert the date column to date (dt.tz_localize is used to remove toe T-Z string)
    game_data['date'] = pd.to_datetime(game_data['date'], errors='coerce').dt.tz_localize(None)
        
    return game_data
```

## Creating function to shape dataframe

id, East Score, West Score

```python
def games_data_reshaped(data):
    
    data['east_team'] = np.where(data['home_conference']=="East", data['home_full_name'], data['visitor_full_name'])
    data['east_score'] = np.where(data['home_conference']=="East", data['home_team_score'], data['visitor_team_score'])
    
    data['west_team'] = np.where(data['home_conference']=="West", data['home_full_name'], data['visitor_full_name'])
    data['west_score'] = np.where(data['home_conference']=="West", data['home_team_score'], data['visitor_team_score'])
    
    data['west_minus_east'] = data['west_score'] - data['east_score']
    
    data = data[['id', 'date', 'season', 'east_team', 'east_score', 'west_team', 'west_score', 'west_minus_east']]
    return data
```

## Create regression function

```python
X = games_data_reshaped['east_score']
y = games_data_reshaped['west_score']
X = sm.add_constant(X)
mod = sm.OLS(y, X, hasconst= True)
res = mod.fit()
print(res.summary())
```

# Stats Data

## Creating function to get data

```python
def get_stats_data(per_page=None, total_pages=None):

    # Get the last page from the meta data
    url = 'https://www.balldontlie.io/api/v1/stats'
    parameters = {'per_page': per_page, 'page': 1}
    resp = requests.get(url, params = parameters).json()
    total_pages = min(x for x in [resp['meta']['total_pages'], total_pages] if x is not None)
    
    # Initialize data frame 
    stats_data = pd.DataFrame()
    
    
    # Loop to iterate through all pages
    for page in range(1, total_pages+1):
        
        # Specifyig parameters
        per_page = min(x for x in [per_page, 100] if x is not None)
        parameters = {'per_page': per_page, 'page': page}
        
        # Making the request
        resp = requests.get(url, params = parameters).json()

        # Creating dataframe from the data key
        data = pd.DataFrame.from_records(resp['data'])
               
        # Creating columns out of the dictionary-type columns
        for col in data.columns:
            if type(data[col][0]) == dict:
                for k in data[col][0].keys():
                    data[col+"_"+k] = data[col].apply(lambda x: x[k])
        
       
        # Column to display which page results came from
        data['page'] = page
        
        # Concatenate our initial datafram with dataframe from each loop
        stats_data = pd.concat([stats_data, data])
        
        # Delay between requests
        time.sleep(0.8)
    
        
    return stats_data
```

```python
def get_data(data, per_page=None, total_pages=None):
    
    # Get the last page from the meta data
    parameters = {'per_page': 1, 'page': 1}
    resp = requests.get(data_dict[data]['url'], params = parameters).json()
    total_pages = min(x for x in [resp['meta']['total_pages'], total_pages] if x is not None)
    
    # Initialize data frame 
    initial_df = pd.DataFrame()
    
    
    # Loop to iterate through all pages
    for page in range(1, total_pages+1):
        
        print(data, page, total_pages)
        start_time = time.time()

        
        # Making the request
        per_page = min(x for x in [data_dict[data]['per_page'], per_page] if x is not None)
        parameters = {'per_page': per_page, 'page': page}
        resp = requests.get(data_dict[data]['url'], params = parameters).json()

        # Creating dataframe from the data key
        df = pd.DataFrame.from_records(resp['data'])
        

        # Creating columns out of the dictionary-type values
        for col in  df.columns:
            if type(df[col][0]) == dict:
                for k in df[col][0].keys():
                    df[col+"_"+k] = df[col].apply(lambda x: x[k])
        
        # Column to display which page results came from
        df['page'] = page
        
        # Concatenate our initial datafram with dataframe from each loop
        initial_df = pd.concat([initial_df, df])
        
        end_time = time.time()
        total_time = end_time - start_time
        
        # Delay between requests
        time.sleep(0.8)
            
    return initial_df
```
